```python
import ast
```

```python
def convert(obj):
  L = []
  for i in ast.literal_eval(obj):
      L.append(i['name'])
  return L
```

```python
# Only apply the convert function if the 'genres' column still contains string data that needs parsing.
# We check if any element in the 'genres' column is a string.
if movies['genres'].apply(lambda x: isinstance(x, str)).any():
    movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
def convert_3(obj):
  L = []
  counter = 0
  for i in ast.literal_eval(obj):
    if counter != 3:
      L.append(i['name'])
      counter+=1
    else:
        break
  return L
```

```python
movies['cast'] = movies['cast'].apply(convert_3)
```

```python
movies.head()
```

```python
def fetch_director(obj):
  L = []
  for i in ast.literal_eval(obj):
    if i['job'] == 'Director':
      L.append(i['name'])
      break
  return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies['overview'][0]
```

```python
movies['overview']= movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies['genres']   = movies['genres'].apply(lambda x:[i.replace("","") for i in x ])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace("","") for i in x ])
movies['cast']     = movies['cast'].apply(lambda x:[i.replace("","") for i in x ])
movies['crew']     = movies['crew'].apply(lambda x:[i.replace("","") for i in x ])
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
new_df = movies[['movie_id','title','tags']].copy()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
new_df['tags'][0]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df['tags']
```

```python
new_df.head()
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
  y=[]

  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
# BAG OF WORDS (VECTORIZATION OF TEXT-NUMERICAL VALUES)
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
'''STEMMING (CONVERTING DIFF. WRDS WITH SAME MEANING INTO SAME WORDS)
['LOVED','LOVING','LOVE'] --> INTO --> ['LOVE','LOVE','LOVE']
'''
```

```python
# TO FIND THE RELEVENCE
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])),reverse = True,key = lambda x:x[1]) # FINDING THE DISTANCE BTW THE 0th MOVIE TO THE OTHER MOVIES
```

```python
def recomend(movie):
  movie_index = new_df[new_df['title'] == movie].index[0]
  distances = similarity[movie_index]
  movies_list = sorted(list(enumerate(distances)), reverse = True, key = lambda x:x[1])[1:6]

  for i in movies_list:
    print(new_df.iloc[i[0]].title)
```

```python
recomend('Batman Begins')
```
